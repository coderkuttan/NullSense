## Step 4 — Set up data.yaml with correct paths

````python
import yaml, glob, os

# Auto-detect the dataset root (handles nested zip structures)
yaml_files = glob.glob('/content/pothole/**/data.yaml', recursive=True)
if not yaml_files:
    raise FileNotFoundError('data.yaml not found! Check your zip structure.')

dataset_root = os.path.dirname(yaml_files[0])
print(f'Dataset root: {dataset_root}')

# Write fixed data.yaml with absolute paths
fixed_yaml = '/content/pothole_data.yaml'
data = {
    'train': os.path.join(dataset_root, 'train', 'images'),
    'val':   os.path.join(dataset_root, 'valid', 'images'),
    'test':  os.path.join(dataset_root, 'test',  'images'),
    'nc':    1,
    'names': ['pothole']
}

with open(fixed_yaml, 'w') as f:
    yaml.dump(data, f, default_flow_style=False)

print(f'data.yaml written:')
print(f'  train: {data["train"]}')
print(f'  val  : {data["val"]}')
print(f'  nc   : {data["nc"]}  classes: {data["names"]}')

# Count images
for split in ['train', 'valid', 'test']:
    img_dir = os.path.join(dataset_root, split, 'images')
    if os.path.exists(img_dir):
        n = len([f for f in os.listdir(img_dir) if f.lower().endswith(('.jpg','.jpeg','.png'))])
        print(f'  {split}: {n} images')
````

## Step 5 — Fine-tune YOLOv8n on potholes

This starts from the pretrained `yolov8n.pt` (80 COCO classes) and adds pothole detection.

**Settings:**
- `epochs=50` — good balance of speed and accuracy
- `batch=16` — fits T4 GPU
- `patience=10` — stops early if no improvement
- `pretrained=True` — keeps all original COCO class weights

````python
from ultralytics import YOLO
import torch

device = 0 if torch.cuda.is_available() else 'cpu'
print(f'Training on: {"GPU" if device == 0 else "CPU"}')

model = YOLO('yolov8n.pt')  # downloads pretrained weights automatically

results = model.train(
    data       = '/content/pothole_data.yaml',
    epochs     = 50,
    imgsz      = 640,
    batch      = 16,
    name       = 'nullsense_pothole',
    project    = '/content/runs',
    pretrained = True,
    patience   = 10,
    save       = True,
    device     = device,
    verbose    = True,
    plots      = True,
)

print('Training complete!')
````

## Step 6 — Validate the trained model

````python
import glob

# Find the best model (handles run numbering)
candidates = glob.glob('/content/runs/nullsense_pothole*/weights/best.pt')
best_pt = sorted(candidates)[-1] if candidates else None

if not best_pt:
    raise FileNotFoundError('best.pt not found! Did training complete?')

print(f'Best model: {best_pt}')

model = YOLO(best_pt)

print('\nModel classes:')
for i, name in model.names.items():
    print(f'  {i}: {name}')

# Validate on val set
metrics = model.val(data='/content/pothole_data.yaml', verbose=False)
map50   = metrics.box.map50
map5095 = metrics.box.map

target = 0.65
status = 'MEETS target' if map50 >= target else 'BELOW target (consider more epochs)'
print(f'\nmAP50   : {map50:.3f}  — {status}')
print(f'mAP50-95: {map5095:.3f}')
````

## Step 7 — Preview detections on sample images

````python
import glob, random
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

# Pick 4 random test images
test_imgs = glob.glob(os.path.join(dataset_root, 'test', 'images', '*.jpg'))
samples   = random.sample(test_imgs, min(4, len(test_imgs)))

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

for ax, img_path in zip(axes, samples):
    results = model(img_path, conf=0.4, verbose=False)
    annotated = results[0].plot()
    ax.imshow(annotated[:, :, ::-1])
    ax.set_title(os.path.basename(img_path), fontsize=8)
    ax.axis('off')

plt.suptitle('NullSense Pothole Detection — Sample Results', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('/content/sample_detections.png', dpi=150, bbox_inches='tight')
plt.show()
print('Saved preview to /content/sample_detections.png')
````
